# Asteroids Data

The first data consists of asteroids which are in watch because of possible impacts on Earth. We used a csv file from https://www.kaggle.com/nasa/asteroid-impacts/version/1#impacts.csv and uploaded the csv file into jupyter notebook in order to construct a dataframe useful for plotting. 

```python
# Reading the asteroids csv data file 
impacts = pd.read_csv(file1)

# Displaying the data table for preview
impacts.head()
```

```python
asteroids = impacts.rename(columns ={"Object Name": "Name"})
asteroids.head()
```

```python
asteroids.loc[asteroids['Maximum Palermo Scale'].idxmax()]
```

```python
def impact_score(impact):
    if impact < -8:
        return 'darkmagenta'
    elif impact < -4:
        return 'LightBlue'
    else:
        return 'red'
    
def asteroid_scale(scale):
    return (scale) * -100;
  

x = asteroids['Possible Impacts']
y = asteroids['Period Start']
z = asteroids['Maximum Palermo Scale']

asteroids_list = asteroids['Maximum Palermo Scale']
asteroids_list_colors = [impact_score(impact) for impact in asteroids_list]

asteroids_scale_list= [asteroid_scale(scale) for scale in asteroids_list]

fig = plt.figure(figsize=(10,8))
plt.ylim(2016,2030)
plt.xlim(-10,100)
plt.scatter(x, y, s=asteroids_scale_list , c=asteroids_list_colors, alpha=0.6, marker="o", edgecolors="black")

plt.title("Possible Future Impacts")
plt.xlabel("Possible Impacts")
plt.ylabel("Years")
plt.grid(True)

plt.savefig("../possible_impacts_asteroids.png")
plt.show() 
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.scatter(asteroids["Possible Impacts"]/(asteroids['Period End'] - asteroids['Period Start']), 
           asteroids["Asteroid Velocity"],s=asteroids["Asteroid Diameter (km)"]*1000, alpha=0.5)
ax.set_xscale('log')
ax.set_xlim([1e-2,20])
ax.set_xlabel('Number of possible impacts per year')
ax.set_ylabel('Asteroid speed (km/s)')
plt.title("Impacts by Speed")
plt.grid(True)
sns.set(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, 
        color_codes=False, rc=None)
plt.savefig("../possible_impacts_asteroids.png")
```

## Meteorites Data

This data consists of meteorites which have struck Earth. We used a csv file from https://www.kaggle.com/nasa/meteorite-landings and uploaded the csv file into jupyter notebook in order to construct a dataframe useful for plotting.

```python
# Reading the meteor csv data file 
meteor = pd.read_csv(file2)
meteor.head()
```

```python
meteor = meteor.rename(columns ={"recclass": "type",
                                 "fall": "discovery",
                                 "nametype": "appearance",
                                 "reclat": "lat",
                                 "reclong": "long",
                                 "GeoLocation": "coordinates"})
meteor.head()
```

```python
#Dropping some unecessary columns 
meteor = meteor.drop(columns = ["id", "coordinates"])
meteor.head()
```

```python
# To sort from highest to lowest, ascending=False must be passed in
meteor_df = meteor.sort_values("year", ascending=False)
meteor_df.head()
```

```python
meteor_count = meteor["year"].value_counts()
meteor_count.head()
```

```python
# Remove the rows with missing data
clean_meteor_df = meteor.dropna(how="any")
clean_meteor_df.count()
```

```python
clean_meteor_df.head()
```

```python
def get_discovery_color(discovery):
    if discovery == 'Fell':
        return 'Green'
    elif discovery == 'Found':
        return 'Red'
    else:
        return 'Gold'

lat = clean_meteor_df['lat'].values
lon = clean_meteor_df['long'].values
mass = clean_meteor_df['mass'].values

discovery_type_list = clean_meteor_df['discovery']
discovery_color_list = [get_discovery_color(discovery_type) for discovery_type in discovery_type_list]

fig = plt.figure(figsize=(16,11))
m = Basemap()
m.shadedrelief()

m.scatter(lon, lat, latlon=True,
          c=discovery_color_list, s=mass/20000,
          alpha=0.5, marker="o", edgecolors="black")

plt.title("Meteorites Impacts by Mass")
plt.savefig("../meteorites_impact_by_mass.png")
```
