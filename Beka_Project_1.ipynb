```python
#Importing the dependecies 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
from citipy import citipy

```

```python
# Loading the file
file1 = "impacts.csv"
file2 = "meteorite-landings.csv"

# Reading the meteor csv data file 
impacts = pd.read_csv(file1)

# Displaying the data table for preview
impacts.head(5)

```

```python
asteroids = impacts.rename(columns ={"Object Name": "Name"})
asteroids.head()
```

```python
# Using GroupBy in order to separate the data into fields according to "period start" values
grouped_asteroids_df = asteroids.groupby(['Period Start'])['Possible Impacts'].count()
grouped_asteroids_df.head()
```
